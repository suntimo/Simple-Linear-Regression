# Simple Linear Regression

```python
#importing the importing library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the datasets
df = pd.read_excel('ML/slr.xls')
```

```python
df.head(3)
```

```python
#obtaining the independent and dependent variables 
x = df.iloc[:,:1].values
y = df.iloc[:,1:].values
```

```python
#show if the variable follows a linear pattern
plt.plot(x,y)
```

```python
sns.relplot(x = 'X', y = 'Y', data = df)
```

```python
sns.scatterplot(x = 'X', y = 'Y', data = df)
```

```python
#spliting the datasets into test and train
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 5) 
```

```python
x_train
```

```python
#importing the linear model
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
#fitting the regression model
reg = regression.fit(x_train, y_train)
```

```python
reg
```

```python
#predicting the value of depedent variable from the independent variable
y_pred = reg.predict(x_test)
```

```python
#Predicted values of dependent variable
y_pred
```

```python
#The true values of dependent variable
y_test
```

```python
#Getting the y_intercept 
reg.intercept_
```

```python
#Getting the coefficient
reg.coef_
```

## Model Evaluation

```python
# import metric library
from sklearn import metrics
```

```python
# mean absolute error (MAE)
MAE = metrics.mean_absolute_error(y_test, y_pred)
```

```python
MAE
```

```python
# Mean Squared Error (MSE)
MSE = metrics.mean_squared_error(y_test, y_pred)
```

```python
MSE
```

```python
# Root mean squared error 
RMSE = np.sqrt(MSE)
```

```python
RMSE
```
